```python
import pandas as pd
import datetime
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
#Graphs
df_graph_data = pd.read_csv('/content/Updated_Inclusive_Indian_Online_Scam_Dataset (1).csv', parse_dates= True, low_memory = False)
```

```python
#Selecting Needed Columns
df1_graph_data = df_graph_data[['location','card_type' , 'customer_age', 'fraud_type', 'amount']]
```

```python
# Overview card_type vs fraud_type

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['fraud_type'].value_counts()
    for x_label, grp in df1_graph_data.groupby('card_type')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('card_type')
_ = plt.ylabel('fraud_type')
```

```python
value_counts = df1_graph_data['location'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Location Attacks')
plt.show()
```

```python
df_grouped = df1_graph_data.groupby(['location', 'card_type']).size().unstack(fill_value=0)

# Plotting
ax = df_grouped.plot(kind='bar', figsize=(14, 7), width=0.4)

# Adding titles and labels
plt.title('Frequency of Card Types by Location')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.legend(title='Card Type')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
df_grouped = df1_graph_data.groupby(['location', 'fraud_type']).size().unstack(fill_value=0)

# Plotting
ax = df_grouped.plot(kind='bar', figsize=(14, 7), width=0.6)

# Adding titles and labels
plt.title('Frequency of Card Types by Location')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.legend(title='Card Type')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

# Histogram for 'location'
plt.subplot(1, 2, 1)
df1_graph_data['location'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Frequency of Locations')
plt.xlabel('Location')
plt.ylabel('Frequency')

# Histogram for 'card type'
plt.subplot(1, 2, 2)
df1_graph_data['card_type'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Frequency of Card Types')
plt.xlabel('Card Type')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

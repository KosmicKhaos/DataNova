```python
import pandas as pd
import datetime
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
#Graphs
df_graph_data = pd.read_csv('/content/Updated_Inclusive_Indian_Online_Scam_Dataset (1).csv', parse_dates= True, low_memory = False)
```

```python
#Selecting Needed Columns
df1_graph_data = df_graph_data[['location','card_type' , 'customer_age', 'fraud_type', 'amount']]
```

Mumbai Analysis

```python
dfMumbai_graph_data = df1_graph_data[((df1_graph_data['location']== "Mumbai"))]
```

```python
# Mumbai card_type

from matplotlib import pyplot as plt
import seaborn as sns
dfMumbai_graph_data.groupby('card_type').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
#Mumbai fraud_type

from matplotlib import pyplot as plt
import seaborn as sns
dfMumbai_graph_data.groupby('fraud_type').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Mumbai amount

from matplotlib import pyplot as plt
dfMumbai_graph_data['amount'].plot(kind='hist', bins=20, title='amount')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
